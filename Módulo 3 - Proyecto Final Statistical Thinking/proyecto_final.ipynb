## Hipótesis 1: El precio varia en función de las noches que el cliente reserva. Cuantas más noches seguidas reserve, el precio/noche disminuye

#### Preparamos los datos según las necesidades de nuestra hipotesis

```python
# Primero definimos los intervalos para agrupar las noches
bins = [1, 3, 5, 7]  # Nuestros intervalos son los siguientes: intervalos [1,3), [3,5), [5,7)
labels = ['1-3 noches', '3-5 noches', '5-7 noches']  # Creamos las etiquetas para los intervalos de pernoctaciones

# Finalmente creamos la columna 'intervalo' en el DF para poder trabajar con ellos 
df_final.loc[:, 'intervalo'] = pd.cut(df_final['minimum_nights'], bins=bins, right=False, labels=labels)

# Preparamos nuestro DF para empexar a trabajar con él 
## Quitamos los nulos
df_final = df_final.dropna()

## Creamos la media de precio por noche según nuestro intervalo y la añadimos a nuestro DF
precios_por_intervalo = df_final.groupby('intervalo', observed=False)['price'].mean()
```

### Creamos un boxplot para comparar los quantiles respecto el intervalo de pernoctaciones

```python
# Creamos el gráfico de boxplot
plt.figure(figsize=(6, 4))
sns.boxplot(x='intervalo', y='price', data=df_final, showfliers=False)  # Eliminamos outliers puestp que sino no podemos visualizar correctamente las cajas del gráfico, ya que había outliers muy altos.

# Añadimos las medianas calculadas
for intervalo, media in precios_por_intervalo.items():
    plt.scatter(intervalo, media, color='red', label='Media' if intervalo == labels[0] else "", zorder=5)

# Ajustamos el diseño del gráfico
plt.title("Distribución del precio por noche según intervalos de noches reservadas")
plt.xlabel("Intervalo de noches reservadas")
plt.ylabel("Precio por noche")
plt.legend()
plt.show()
```

### Creamos un gráfico "barplot" para ver la diferencia de precio según los 3 intervalos de pernoctaciones detallados en el roadmap: 

```python
# Creamos el gráfico
plt.bar(precios_por_intervalo.index, precios_por_intervalo, edgecolor='black')
plt.title("Precio promedio por noche según intervalos de noches seguidas reservadas")
plt.xlabel("Intervalo de noches reservadas")
plt.ylabel("Precio promedio por noche")
plt.show()
```

### Creamos un Scatter plot con línias de tendencias

```python
# Agrupamos por intervalos y calculamos el precio promedio
data_scatter = df_final.groupby('intervalo', observed=False)['price'].mean().reset_index()

# Crear el scatter plot con línea de tendencia
plt.figure(figsize=(8, 6))
sns.scatterplot(x='intervalo', y='price', data=data_scatter, s=100, color='blue', label='Datos')
sns.lineplot(x='intervalo', y='price', data=data_scatter, color='red', label='Tendencia')

plt.title("Relación entre intervalos de noches y precio promedio")
plt.xlabel("Intervalos de noches")
plt.ylabel("Precio promedio por noche")
plt.legend()
plt.grid(True)
plt.show()
```

## Hipótesis 2: El precio varia en función de la capacidad y las habitaciones del inmueble

### Capacidad 4pax: apartamento con 1 habitación vs 2 habitaciones

```python
# Crear una columna que describa la distribución de habitaciones y capacidad
def categorize_configuration(row):
    if row['accommodates'] == 4:
        if row['bedrooms'] == 1:
            return "4pax: 1 hab + sofá-cama"
        elif row['bedrooms'] == 2:
            return "4pax: 2 hab"
    elif row['accommodates'] == 6:
        if row['bedrooms'] == 2:
            return "6pax: 2 hab + sofá-cama"
        elif row['bedrooms'] == 3:
            return "6pax: 3 hab"
    return "Other"

df_final['tipologia'] = df_final.apply(categorize_configuration, axis=1)
```

```python
# Filtrar las configuraciones relevantes
df_final_filtrado = df_final[df_final['tipologia'].isin([
    "4pax: 1 hab + sofá-cama",
    "4pax: 2 hab",
    "6pax: 2 hab + sofá-cama",
    "6pax: 3 hab"
])]

# Crear el gráfico de barras
plt.figure(figsize=(10, 6))
sns.barplot(
    data=df_final_filtrado, 
    x="tipologia", 
    y="price", 
    estimator="mean", 
    errorbar='sd',  # Incluye barras de error para la desviación estándar
)
plt.title("Precio promedio por la capacidad y la tipologia del inmueble")
plt.xlabel("Tipologia")
plt.ylabel("Precio promedio")
plt.xticks(rotation=15)
plt.show()
```

```python
# Calculamos los percentiles 25 (Q1) y 75 (Q3)
Q1 = df_final_filtrado['price'].quantile(0.25)
Q3 = df_final_filtrado['price'].quantile(0.75)

# Calculamos el rango intercuartílico 
rango_interquantil = Q3 - Q1

# Definir los límites inferior y superior para los outliers
lower_bound = Q1 - 1.5 * rango_interquantil
upper_bound = Q3 + 1.5 * rango_interquantil

# Filtrar los datos para eliminar los outliers
df_final_filtrado_sin_outliers = df_final_filtrado[(df_final_filtrado['price'] >= lower_bound) & (df_final_filtrado['price'] <= upper_bound)]

# Violin Plot sin outliers
plt.figure(figsize=(12, 6))
sns.violinplot(
    data=df_final_filtrado_sin_outliers,
    x="tipologia",
    y="price",
)
plt.title("Distribución de precios por configuración de capacidad y habitaciones (sin outliers)")
plt.xlabel("Tipologia")
plt.ylabel("Precio")
plt.xticks(rotation=15)
plt.show()

# Boxplot sin outliers
plt.figure(figsize=(12, 6))
sns.boxplot(
    data=df_final_filtrado_sin_outliers,
    x="tipologia",
    y="price",
)
plt.title("Distribución de precios por configuración de capacidad y habitaciones (sin outliers)")
plt.xlabel("Configuración")
plt.ylabel("Precio")
plt.xticks(rotation=15)
plt.show()
```

## Hipótesis 3: La ubicación influye en el precio: cuanto más cerca de sol, más alto es el precio/noche
    - Distancia 1: a 1km de sol 
    - Distancia 2: a 5km de sol
    - Distancia 3: a 10km de sol
    - Distancia 4: a 30km de sol 

```python
# Crear los intervalos de distancia
bins = [0, 1, 2, 3, 4, 5, 8, 10, 15]  # Bordes de los intervalos
labels = ['1km', '2km','3km','4km','5km','8km','10km','15km']  # Etiquetas de los intervalos

# Crear la columna 'dist_interval' correctamente
df_final['dist_interval'] = pd.cut(
    df_final['dist_km_sol'], 
    bins=bins, 
    labels=labels, 
    right=True
)
# Filtrar los precios dentro del rango permitido
filtered_df = df_final[(df_final['price'] >= lower_bound) & (df_final['price'] <= upper_bound)]

# Agrupar por intervalo de distancia y calcular el precio promedio
mean_prices = df_final.groupby('dist_interval', observed=False)['price'].mean().reset_index()

# Ordenar los intervalos
mean_prices['dist_interval'] = pd.Categorical(mean_prices['dist_interval'], 
                                              categories=['1km', '2km','3km','4km','5km','8km','10km','15km'], 
                                              ordered=True)
mean_prices = mean_prices.sort_values('dist_interval')
```

```python
# Crear el bar plot
plt.figure(figsize=(10, 6))
sns.barplot(data=mean_prices, x='dist_interval', y='price')
plt.title('Precio promedio por intervalo de distancia a Sol', fontsize=16)
plt.xlabel('Intervalo de distancia a Sol', fontsize=14)
plt.ylabel('Precio promedio (€/noche)', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='dist_interval', y='price', data=filtered_df, order=labels)
plt.title('Distribución de precios por distancia al Sol (sin outliers)')
plt.xlabel('Distancia al Sol')
plt.ylabel('Precio por noche')
plt.xticks(rotation=45)
plt.show()
```

```python
# Primero, podemos asegurarnos de que las políticas de cancelación sean categóricas ordenadas
cancelation_order = ['strict_14_with_grace_period', 'moderate', 'flexible', 'super_strict_30', 'super_strict_60']
df_final['cancellation_policy'] = pd.Categorical(df_final['cancellation_policy'], categories=cancelation_order, ordered=True)

# 1. Eliminar los outliers de 'price' usando el IQR
Q1 = df_final['price'].quantile(0.25)
Q3 = df_final['price'].quantile(0.75)
IQR = Q3 - Q1

# Filtramos los precios que están fuera del rango [Q1 - 1.5*IQR, Q3 + 1.5*IQR]
df_no_outliers = df_final[(df_final['price'] >= (Q1 - 1.5 * IQR)) & (df_final['price'] <= (Q3 + 1.5 * IQR))]

# 2. Boxplot sin outliers: Comparar la distribución de precios por política de cancelación
plt.figure(figsize=(10, 6))
sns.boxplot(x='cancellation_policy', y='price', data=df_no_outliers, order=cancelation_order)
plt.title('Distribución de precios por política de cancelación (sin outliers)')
plt.xlabel('Política de cancelación')
plt.ylabel('Precio por noche')
plt.xticks(rotation=45)
plt.show()

# 3. Bar Plot sin outliers: Mostrar el precio promedio por política de cancelación
plt.figure(figsize=(10, 6))
mean_prices_no_outliers = df_no_outliers.groupby('cancellation_policy', observed=False)['price'].mean().reset_index()
sns.barplot(x='cancellation_policy', y='price', data=mean_prices_no_outliers, order=cancelation_order)
plt.title('Precio promedio por política de cancelación (sin outliers)')
plt.xlabel('Política de cancelación')
plt.ylabel('Precio promedio por noche')
plt.xticks(rotation=45)
plt.show()
```
